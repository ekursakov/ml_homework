Классификатор с 2 скрытыми слоями показывает немного лучший результат, поэтому далее используем 2 скрытых слоя

**Подбор alpha**

```python
alpha_arr = np.logspace(-3, 2, 21)
test_err = []
train_err = []
for alpha in alpha_arr:
    mlp_model = MLPClassifier(alpha = alpha, hidden_layer_sizes = (100,100), 
                              solver = 'lbfgs', activation = 'logistic', random_state = 42)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
plt.semilogx(alpha_arr, test_err, 'b-o', label = 'test')
plt.semilogx(alpha_arr, train_err, 'r-o', label = 'train')
plt.xlim([np.max(alpha_arr), np.min(alpha_arr)])
plt.title('Error vs. alpha')
plt.xlabel('alpha')
plt.ylabel('error')
plt.legend()
pass
```

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)
```

**random forests**

```python
from sklearn import ensemble
rndforest = ensemble.RandomForestClassifier(n_estimators = 500, oob_score=True)
rndforest.fit(X_train, y_train)

err_train = np.mean(y_train != rndforest.predict(X_train))
err_test = np.mean(y_test != rndforest.predict(X_test))
oob_error = 1 - rndforest.oob_score_
print(err_train, err_test, oob_error)
```

```python
tree_counts = (50, 100, 200, 500, 1000, 2000, 5000)
test_err = []
train_err = []
oob_error = []

for count in tree_counts:
    rf = ensemble.RandomForestClassifier(n_estimators = count, oob_score=True)
    rf.fit(X_train, y_train)

    y_train_pred = rf.predict(X_train)
    y_test_pred = rf.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
    oob_error.append(1 - rf.oob_score_)
```

```python
plt.semilogx(tree_counts, train_err, 'b-o', label = 'train')
plt.semilogx(tree_counts, test_err, 'r-o', label = 'test')
plt.semilogx(tree_counts, oob_error, 'g-o', label = 'oob')
plt.xlim([np.min(tree_counts), np.max(tree_counts)])
plt.title('Random Forest: Error vs. Trees')
plt.xlabel('trees')
plt.ylabel('error')
plt.legend()
pass
```

# ДЗ №7: PCA

Исходные данные содержат 58 столбцов, уменьшим до 30.

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components = 30)
X_train_r = pca.fit_transform(X_train)
X_test_r = pca.transform(X_test)
```

```python
from sklearn import ensemble
cls = ensemble.AdaBoostClassifier(n_estimators = 100)
cls.fit(X_train_r, y_train)
```

```python
err_train = np.mean(y_train != cls.predict(X_train_r))
err_test  = np.mean(y_test != cls.predict(X_test_r))
print(err_train, err_test)
```

```python
compoments_count = np.arange(5, 58, 5)
test_err = []
train_err = []

for count in compoments_count:
    pca = decomposition.PCA(n_components = count)
    X_train_r = pca.fit_transform(X_train)
    X_test_r = pca.transform(X_test)
    cls = ensemble.AdaBoostClassifier(n_estimators = 100)
    cls.fit(X_train_r, y_train)
    train_err.append(np.mean(y_train != cls.predict(X_train_r)))
    test_err.append(np.mean(y_test != cls.predict(X_test_r)))
    print(count)
```

```python
plt.plot(compoments_count, train_err, 'b-o', label = 'train')
plt.plot(compoments_count, test_err, 'r-o', label = 'test')
plt.xlim([np.max(compoments_count), np.min(compoments_count)])
plt.title('Error vs. components count')
plt.xlabel('components')
plt.ylabel('error')
plt.legend()
pass
```
